```python
import pandas as pd
from IPython.display import display
# Membaca dataset iris
data = pd.read_csv('iris.csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

# Mengkonversi kolom Class ke numerik 
data['species'] = pd.Categorical(data['species']).codes

# Memisahkan fitur numerik
numeric_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Menghitung statistik deskriptif
min_values = data[numeric_columns].min()
max_values = data[numeric_columns].max()
mean_values = data[numeric_columns].mean()
std_dev_values = data[numeric_columns].std()

# Menghitung korelasi dengan Class
correlations = []
for column in numeric_columns:
    correlation = np.corrcoef(data[column], data['species'])[0,1]
    correlations.append(correlation)

# Membuat DataFrame dengan statistik
stats_data = {
    'Min': min_values,
    'Max': max_values,
    'Mean': mean_values,
    'Std Deviasi': std_dev_values,
    'Class Correlation': correlations
}

# Membuat DataFrame
stats_df = pd.DataFrame(stats_data, index=numeric_columns)

# Menampilkan hasil
print("\nPenghitungan Statistika Min, Max, Std Deviasi, serta Class Korelasi(Correlation):")

display(stats_df)
```

```python
# Mengimport libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat sample data
data = pd.DataFrame({
    'species': ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
    'sepal_length': [250, 300, 325]  # Nilai yang disesuaikan dengan gambar
})

# Set style untuk plot yang lebih bersih
plt.style.use('classic')

# Membuat color diagram
colors = ['#6FB1E4', '#40D9CF', '#4682B4']  # Light blue, turquoise, steel blue

# Membuat bar chart
plt.figure(figsize=(10, 7))
sns.barplot(x='species', y='sepal_length', data=data, 
            palette=colors, ci=None)

# Membuat Plot
plt.title('Bar Chart', pad=15, fontsize=12)
plt.xlabel('Species', fontsize=10)
plt.ylabel('sepal length', fontsize=10)

# Customize y-axis
plt.ylim(0, 350)
plt.yticks([0, 50, 100, 150, 200, 250, 300, 350])

# Menambahkan grid lines untuk y-axis
plt.grid(axis='y', linestyle='-', alpha=0.2)

# Menghapus spine atas dan kanan
sns.despine()

# Membuat spine kiri dan bawah berwarna hitam
plt.gca().spines['left'].set_color('black')
plt.gca().spines['bottom'].set_color('black')

# Membuat tick marks lebih panjang
plt.tick_params(axis='both', which='major', length=6)

# Menyesuaikan layout
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset iris
data = pd.read_csv('iris.csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

# Membuat histogram untuk kolom 'sepal_length'
plt.figure(figsize=(8, 6))
plt.hist(data['sepal_length'], bins=10, edgecolor='black')
plt.title('Sepal Length')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

# Menampilkan plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Memuat dataset Iris
iris = sns.load_dataset('iris')

# Mengatur gaya untuk tampilan yang lebih bersih
sns.set_style("whitegrid")

# Membuat plot sebaran panjang sepal vs lebar sepal, diwarnai berdasarkan spesies
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, 
                s=100, edgecolor='black', palette='Dark2', alpha=0.7)

# Menyesuaikan plot
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Menampilkan plot
plt.show()
```
